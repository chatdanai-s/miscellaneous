```python
# Generate 50 points from f(x) = sin(2*pi*x)exp(-2x) where x in [0,1]
# Then use MLP to approximate f(x) with the universal approximation theorem
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import Sequential, layers
```

```python
def f(x):
    out = np.sin(2*np.pi*x) * np.exp(-2*x)
    return out
x = np.linspace(0,1,50); y = f(x)
plt.plot(x,y,'.')
```

```python
def createModel():
    model = Sequential()
    model.add(layers.Input(shape=(1,)))
    model.add(layers.Dense(10, activation='selu'))
    model.add(layers.Dense(10, activation='selu'))
    model.add(layers.Dense(10, activation='selu'))
    model.add(layers.Dense(1))
    model.compile(loss='mse', optimizer='adam', metrics=['mae'])
    return model
model = createModel()
h = model.fit(x, y, epochs=3000, verbose=0)
```

```python
yp = model(x)
plt.plot(x,y,'.', x,yp,'-')
```

```python
plt.plot(h.epoch, h.history['loss'])
plt.yscale('log')
```
