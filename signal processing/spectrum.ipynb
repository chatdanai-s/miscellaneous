```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DTFT
fs = 100
T = 1/fs
f = np.linspace(-0.5*fs, 0.5*fs, fs)
c = 0.5
twopif_fs = 2*np.pi*f*T
A = (1 - 2*c*np.cos(twopif_fs) + c**2)**(-0.5)
phi = -np.arctan((c*np.sin(twopif_fs)) / (1 - c*np.cos(twopif_fs)))

k = np.linspace(0, 10, 100)
x = c**k
plt.plot(k, x)
plt.xlabel('k'); plt.ylabel('x(k) $\mu$(k)')
plt.title(f'Signal x(k) = {c}^k')
plt.xlim(k.min(), k.max())
plt.show()

plt.plot(f*T, A)
plt.xlabel('f/fs'); plt.ylabel('A(f)')
plt.title(f'Amplitude spectrum of x(k) = {c}^k')
plt.xlim(-0.5, 0.5)
plt.show()

plt.plot(f*T, phi)
plt.xlabel('f/fs'); plt.ylabel('$\phi$(f)')
plt.title(f'Phase spectrum of x(k) = {c}^k')
plt.xlim(-0.5, 0.5)
plt.show()
```

```python
# FFT from scipy
from scipy.fft import fft, fftfreq, ifft
x = np.array([1, 2, 3, 4, 5])
y = fft(x)
N = len(x)
f = fftfreq(N, T)
size = 2

plt.scatter(f*T, np.abs(y))
plt.xlabel('f/fs'); plt.ylabel('A(f)')
plt.title(f'Amplitude spectrum of x(k) = {c}^k')
plt.show()

plt.scatter(f*T, np.angle(y))
plt.xlabel('f/fs'); plt.ylabel('$\phi$(f)')
plt.title(f'Phase spectrum of x(k) = {c}^k')
plt.show()

xx = ifft(y)
plt.plot(x, xx)
plt.show()
```

```python
# FFT on sample sinusoidal signal
f = 5
fs = 100
T = 1/fs
t = np.arange(0, 1, T); N = len(t)
x = np.sin(2*np.pi*f*t) + 0.5*np.sin(2*np.pi*2*f*t + 0.5) + np.random.normal(0, 0.3, N)
plt.plot(t, x)
plt.show()

X = fft(x)
f = fftfreq(N, T)

idx = np.argsort(f)
X = X[idx]; f = f[idx]

plt.plot(f, np.abs(X))
plt.show()
```
