```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Window function visualization
L = 250
m = 500
k = np.arange(0, 1000)
rectangle = np.heaviside(k-m, 1) - np.heaviside(k-L-m, 1)
hanning = (0.5 - 0.5 * np.cos(2*np.pi*(k-m)/L)) * rectangle

plt.figure(figsize=(13,5))
plt.plot(k, rectangle, label='Rectangular window')
plt.plot(k, hanning, label='Hanning window')
plt.xlabel('k'); plt.ylabel('w(k)')
plt.xlim(k.min(), k.max())
plt.title('Window functions')
plt.legend()

plt.show()
```

```python
# Noisy spectrum windowed (from denoising.ipynb)
F1 = 100
F2 = 50
fs = 2000
T = 1/fs
signal = np.sin(2 * F1/fs * np.pi * k) + np.cos(2 * F2/fs * np.pi * k)
noise = np.random.normal(loc=0.0, scale=0.8, size=len(signal))

x = signal + noise

plt.figure(figsize=(13, 5))
plt.plot(k, signal, color='black', zorder=1)
plt.plot(k, x, color='red', zorder=0)
plt.title('Noisy signal and theoretical denoised signal (Unwindowed)')
plt.xlabel('k'); plt.ylabel('Signal')
plt.xlim(k.min(), k.max())
plt.show()

plt.figure(figsize=(13, 5))
plt.plot(k, signal * rectangle, color='black', zorder=1)
plt.plot(k, x * rectangle, color='red', zorder=0)
plt.title('Noisy signal and theoretical denoised signal (Rectangular windowed)')
plt.xlabel('k'); plt.ylabel('Signal')
plt.xlim(k.min(), k.max())
plt.show()

plt.figure(figsize=(13, 5))
plt.plot(k, signal * hanning, color='black', zorder=1)
plt.plot(k, x * hanning, color='red', zorder=0)
plt.title('Noisy signal and theoretical denoised signal (Hanning windowed)')
plt.xlabel('k'); plt.ylabel('Signal')
plt.xlim(k.min(), k.max())
plt.show()
```
